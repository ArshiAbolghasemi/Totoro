## 1) OOP core idea

OOP organizes code using **classes** and **objects**:
- **Class**: blueprint
- **Object (instance)**: concrete item created from that blueprint
- **Attributes**: data attached to objects/classes
- **Methods**: functions defined inside a class


## 2) Attributes and methods

In Python, the most important attribute/method categories are:
- **Instance attributes**: unique per object (`self.name`)
- **Class attributes**: shared by all instances
- **Instance methods**: operate on object state (`self`)
- **Class methods**: operate on class state (`cls`)
- **Static methods**: utility logic grouped in class namespace


```python
class BankAccount:
    bank_name = "ABC Bank"  # class attribute

    def __init__(self, owner, balance=0):
        self.owner = owner          # instance attribute
        self.balance = balance      # instance attribute

    def deposit(self, amount):      # instance method
        self.balance += amount
        return self.balance

    @classmethod
    def rename_bank(cls, new_name):
        cls.bank_name = new_name

    @staticmethod
    def is_valid_amount(amount):
        return amount > 0


acc = BankAccount("Aria", 100)
print(acc.owner, acc.balance)
acc.owner = "Ali"
acc.balance = 50
print(acc.owner, acc.balance)
print(BankAccount.bank_name)
print(acc.deposit(50))
print(BankAccount.is_valid_amount(20))
BankAccount.rename_bank("Global Bank")
print(BankAccount.bank_name)
```

## 3) Inheritance + `super()`

**Inheritance** lets a child class reuse and extend parent behavior.

Use `super()` in child classes to call parent methods (especially `__init__`) so parent setup is not duplicated.


```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound"


class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # call parent constructor
        self.breed = breed

    def speak(self):
        base = super().speak()  # call parent method
        return f"{base} -> {self.name} barks"


d = Dog("Max", "Labrador")
print(d.name, d.breed)
print(d.speak())
```

## 4) Special methods (dunder methods)

You are right: methods like `__repr__`, `__len__`, `__enter__`, and `__exit__` are called **special methods** (or magic/dunder methods).

They let your objects work naturally with Python built-ins and syntax:
- `__repr__`: developer-friendly object representation
- `__len__`: enables `len(obj)`
- `__enter__` and `__exit__`: context manager support for `with`


```python
class Report:
    def __init__(self, title, rows):
        self.title = title
        self.rows = rows

    def __repr__(self):
        return f"Report(title={self.title!r}, rows={len(self.rows)})"

    def __len__(self):
        return len(self.rows)


r = Report("Sales", [120, 130, 95])
print(r)          # calls __repr__
print(len(r))     # calls __len__


class FileLogger:
    def __init__(self, path):
        self.path = path
        self.file = None

    def __enter__(self):
        self.file = open(self.path, "w")
        self.file.write("start")
        return self.file

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.file.write("end")
        self.file.close()
        return False


with FileLogger("oop_log.txt") as f:
    f.write("inside with block")
```

## 5) All common inheritance types in Python

Python supports these common inheritance patterns:
1. **Single inheritance**: one child from one parent
2. **Multilevel inheritance**: chain of inheritance (grandparent -> parent -> child)
3. **Multiple inheritance**: one child from multiple parents
4. **Hierarchical inheritance**: multiple children from one parent


```python
# 1) Single inheritance
class Vehicle:
    def move(self):
        return "Vehicle moves"


class Car(Vehicle):
    pass


# 2) Multilevel inheritance
class ElectricCar(Car):
    pass


# 3) Multiple inheritance
class Flyable:
    def fly(self):
        return "Can fly"


class FlyingCar(Car, Flyable):
    pass


# 4) Hierarchical inheritance
class Bike(Vehicle):
    pass


c = Car()
ec = ElectricCar()
fc = FlyingCar()
b = Bike()

print(c.move())
print(ec.move())
print(fc.move(), "|", fc.fly())
print(b.move())
```

## 6) Abstraction

**Abstraction** means exposing what an object does while hiding implementation details.

In Python, a common approach is using abstract base classes (`ABC`) and abstract methods (`@abstractmethod`).
A child class must implement all abstract methods before it can be instantiated.


```python
from abc import ABC, abstractmethod


class PaymentMethod(ABC):
    @abstractmethod
    def pay(self, amount):
        pass


class CardPayment(PaymentMethod):
    def pay(self, amount):
        return f"Paid {amount} with card"


p = CardPayment()
print(p.pay(250))
```
