## 1) Why file handling matters

Files let your programs persist data after execution.
In data workflows, files are used for logs, configuration, exports, and model artifacts.


## 2) Writing to a text file

Use mode `w` to create/overwrite a file.
Use mode `a` to append without removing existing content.


```python
with open('test.txt', 'w') as f:
    f.write('hello file\n')
    f.write('tmep line\n')

print('written')
```

## 3) Reading a text file

`read()` returns the full content as one string.
`readline()` returns one line at a time.


```python
with open('test.txt', 'r') as f:
    content = f.read()

print(content)
```

```python
with open('test.txt', 'r') as f:
    line1 = f.readline()
    line2 = f.readline()

print('line1:', line1.strip())
print('line2:', line2.strip())
```

## 4) Appending to a file

Use `a` or `a+` when you want to keep existing file content and add new lines.


```python
with open('test.txt', 'a') as f:
    f.write('appended line')

with open('test.txt', 'r') as f:
    print(f.read())
```

## 5) Safe pattern with `with open(...)`

`with` automatically closes files, even if an error happens.
This is the recommended default pattern.


```python
# Good pattern
with open('test.txt', 'r') as f:
    first_10_chars = f.read(10)

print(first_10_chars)
```

## 6) common mistakes

### Case A: `w` mode truncates the file
Opening an existing file with `w` erases old content.


```python
with open('test.txt', 'w') as f:
    f.write('only this remains')

with open('test.txt', 'r') as f:
    print(f.read())
```

### Case B: Reading twice from same handle
After one full `read()`, cursor is at end-of-file.
Use `seek(0)` to move back to start.


```python
with open('test.txt', 'r') as f:
    print('first read:', repr(f.read()))
    print('second read:', repr(f.read()))
    f.seek(0)
    print('after seek:', repr(f.read()))
```

### Case C: File might not exist
Handle missing files safely with `try/except`.


```python
filename = 'missing_file.txt'

try:
    with open(filename, 'r') as f:
        print(f.read())
except FileNotFoundError:
    print('File not found:', filename)
```

### Case D: Path separators on Windows
Backslashes can create escape sequences (`
`, `	`).
Use raw strings (`r'...'`) or double backslashes.


```python
print(r'C:\Users\name\data.txt')
print("C:\\Users\\name\\data.txt")
```

## 7) Iterating line by line

Useful for large files where loading everything with `read()` is inefficient.


```python
with open('test.txt', 'r') as f:
    for line_no, line in enumerate(f, start=1):
        print(line_no, line.strip())
```

## 8) Binary files

Use binary modes (`rb`, `wb`) for images, audio, and other non-text data.


```python
# Example: write and read raw bytes
with open('sample.bin', 'wb') as f:
    f.write(b'ABC123')

with open('sample.bin', 'rb') as f:
    data = f.read()

print(data, type(data))
```

## 9) Storing Python objects (`json`, `pickle`, `parquet`)

- `json`: human-readable, interoperable, best for simple data types.
- `pickle`: Python-specific serialization, supports more object types.
- `parquet`: compact columnar format, ideal for tabular analytics data.


```python
import json

record = {'name': 'Sara', 'score': 95, 'skills': ['python', 'sql']}

with open('record.json', 'w') as f:
    json.dump(record, f, indent=2)

with open('record.json', 'r') as f:
    loaded = json.load(f)

print(loaded)
```

```python
import pickle

model_info = {'model': 'baseline', 'accuracy': 0.84}

with open('model.pkl', 'wb') as f:
    pickle.dump(model_info, f)

with open('model.pkl', 'rb') as f:
    loaded_model_info = pickle.load(f)

print(loaded_model_info)
```

### Parquet example

Use Parquet for dataframe-style data when you need efficient storage and fast I/O.


```python
import pandas as pd

df = pd.DataFrame({'name': ['Sara', 'Ali'], 'score': [95, 88]})
df.to_parquet('scores.parquet', index=False)
loaded_df = pd.read_parquet('scores.parquet')
print(loaded_df)
```
