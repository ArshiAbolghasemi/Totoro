## 1) Why Data Types Matter in Data Science

In data science, type mistakes can cause bugs and wrong results.

Examples:
- Numeric columns loaded as strings ("10", "20") cannot be summed correctly.
- Category labels are better represented as strings or encoded integers.
- Missing values must be represented clearly (`None` in pure Python, often `NaN` in pandas).

Let's start with a simple helper to inspect values.

```python
def inspect_value(x):
    print(f"value : {x!r}")
    print(f"type:  {type(x)}")

inspect_value(42)
inspect_value(3.14)
inspect_value("42")
```

## 2) Numeric Types

Python has three main numeric types:
- `int`: whole numbers
- `float`: decimal numbers
- `complex`: numbers with real and imaginary parts (less common in beginner data science)

```python
a = 15          # int
b = 2.5         # float
c = 3 + 4j      # complex

print(type(a), a)
print(type(b), b)
print(type(c), c)
```

```python
# Arithmetic with int and float
x = 10
y = 3

print("x + y =", x + y)
print("x - y =", x - y)
print("x * y =", x * y)
print("x / y =", x / y)   # always float division
print("x // y =", x // y) # floor division
print("x % y =", x % y)   # remainder
print("x ** y =", x ** y) # power
```

### Data science tip: floating-point precision

Floating-point values are stored in binary, so some decimal values are approximate.

```python
print(0.1 + 0.2)
print((0.1 + 0.2) == 0.3)

# Better approach for comparisons
import math
print(math.isclose(0.1 + 0.2, 0.3))
```

### When to use `Decimal` for exact base-10 values

For money and other exact decimal quantities, `float` can introduce tiny rounding errors.
`Decimal` (from Python's built-in `decimal` module) stores decimal numbers more predictably.


```python
from decimal import Decimal


print("float 0.1 + 0.2 ->", 0.1 + 0.2)
print("Decimal('0.1') + Decimal('0.2') ->", Decimal("0.1") + Decimal("0.2"))
```

## 3) Boolean Type (`bool`)

`bool` has two values: `True` and `False`.
It is used in filtering, conditions, and rule-based logic.

```python
age = 21
is_adult = age >= 18

print(is_adult)
print(type(is_adult))

print(True + True)   # bool behaves like 1/0 in arithmetic
print(False * 10)
```

```python
# Useful when filtering data records
temperatures = [36.5, 38.1, 37.0, 39.2]
fever_flags = [t > 37.5 for t in temperatures]
print(fever_flags)
```

## 4) String Type (`str`)

Strings are sequences of text characters.
In data science, they often represent categories, IDs, labels, and raw input values.

```python
name = "Alice"
city = 'New York'

print(name + city)
print(type(name))
print(len(name))
print(name[0], name[-1])
```

```python
# Common string operations
text = "  Python For Data Science  "

print(text.lower())
print(text.upper())
print(text.strip())
print(text.replace("Data Science", "Analytics"))
print("Data" in text)
```

```python
# f-strings for reporting
score = 0.93456
model = "Logistic Regression"
print(f"Model: {model} | Accuracy: {score:.2%}")
```

g## 5) Sequence Types: `list` and `tuple`

### `list`
- Mutable (can change)
- Can store mixed types

### `tuple`
- Immutable (cannot change)
- Often used for fixed records

```python
numbers = [10, 20, 30]
print(numbers)

numbers.append(40)
numbers[0] = 5
print(numbers)
```

```python
point = (3.5, 7.2)  # latitude, longitude for example
print(point)
print(point[0])
```

```python
# Packing and unpacking tuples
record = ("A102", 88.5, True)
student_id, score, passed = record
print(student_id, score, passed)
```

## 6) Set Type (`set`)

A set stores unique, unordered elements.
Useful for removing duplicates and membership testing.

```python
items = ["apple", "banana", "apple", "orange", "banana"]
unique_items = set(items)
print(unique_items)

print("apple" in unique_items)
print("grape" in unique_items)
```

```python
# Set operations
A = {1, 2, 3, 4}
B = {3, 4, 5}

print("union:", A | B)
print("intersection:", A & B)
print("difference A-B:", A - B)
```

## 7) Dictionary Type (`dict`)

Dictionaries store key-value pairs.
They are one of the most important data structures in data science preprocessing.

```python
student = {
    "id": "S001",
    "name": "Sara",
    "age": 22,
    "gpa": 3.8,
    "is_active": True
}

print(student)
print(student["name"])
print(student.get("major", "Not provided"))
```

## 8) Missing Values with `None`

`None` means "no value" or "value not set".
This is common in raw datasets.

```python
income = None
print(income)
print(type(income))
```

## 9) Type Conversion (Casting)

You often need to convert between types when reading files, user input, or APIs.

```python
raw_age = "29"
age = int(raw_age)
print(age, type(age))

raw_price = "19.99"
price = float(raw_price)
print(price, type(price))

score = 95
score_text = str(score)
print(score_text, type(score_text))
```

## 10) Mutability vs Immutability

- **Mutable:** object can be changed after creation (`list`, `dict`, `set`)
- **Immutable:** object cannot be changed (`int`, `float`, `str`, `tuple`, `bool`, `NoneType`)

Understanding this helps prevent accidental bugs.

```python
x = 10
print(id(x))
x = x + 1
print(id(x))
```

```python
# Immutable example (string)
text = "data"
text = text.upper()  # creates a new string
print(text)

# Mutable example (list)
values = [1, 2, 3]
values.append(4)  # modifies same list
print(values)
```

```python
# Why it matters in functions

def add_item_bad(lst, item):
    lst.append(item)
    return lst

original = [1, 2]
updated = add_item_bad(original, 3)

print("original:", original)
print("updated:", updated)
```
