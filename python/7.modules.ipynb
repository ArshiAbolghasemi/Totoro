## 1. Module vs Package <a id="1-module-vs-package"></a>

- A **module** is a single Python file (example: `math_utils.py`).
- A **package** is a folder containing one or more modules (often with `__init__.py`).
- Packages help organize larger projects into logical parts.


````python
# Example module imports from Python standard library
import math
import json

print(math.sqrt(64))
print(json.dumps({"topic": "modules"}))
````

## 2. Import basics <a id="2-import-basics"></a>

You can import full modules, specific functions, or use aliases.


````python
import statistics as stats
from datetime import datetime

values = [10, 20, 30, 40]
print("mean:", stats.mean(values))
print("year:", datetime.now().year)
````

## 3. Installing packages with pip <a id="3-installing-packages-with-pip"></a>

Use these commands in terminal (not inside Python code):

```bash
python -m pip install pandas
python -m pip install "numpy>=1.26,<3"
python -m pip install jupyterlab scikit-learn matplotlib seaborn
```

Install from a requirements file:

```bash
python -m pip install -r requirements.txt
```


## 4. requirements.txt <a id="4-requirementstxt"></a>

`requirements.txt` pins or constrains package versions for reproducible environments.

Example `requirements.txt`:


````python
requirements_txt = """numpy==2.1.1
pandas==2.2.3
matplotlib==3.9.2
seaborn==0.13.2
scikit-learn==1.5.2
jupyterlab==4.2.5
"""

print(requirements_txt)
````

## 5. pyproject.toml <a id="5-pyprojecttoml"></a>

`pyproject.toml` is the modern project configuration file for Python packaging and dependencies.

Minimal example:


````python
pyproject_toml = """[project]
name = "data-science-starter"
version = "0.1.0"
description = "Starter project for data science"
requires-python = ">=3.11"
dependencies = [
  "numpy>=2.1,<3",
  "pandas>=2.2,<3",
  "matplotlib>=3.9,<4",
  "seaborn>=0.13,<1",
  "scikit-learn>=1.5,<2",
  "jupyterlab>=4.2,<5",
]
"""

print(pyproject_toml)
````

## 6. uv workflow <a id="6-uv-workflow"></a>

`uv` is a fast Python package and project manager.

Typical workflow:

```bash
# Create project (creates pyproject.toml)
uv init ds-project
cd ds-project

# Create and use virtual environment
uv venv

# Add dependencies
uv add numpy pandas matplotlib seaborn scikit-learn jupyterlab

# Run python or jupyter in this environment
uv run python -c "import pandas as pd; print(pd.__version__)"
uv run jupyter lab
```


## 7. conda workflow <a id="7-conda-workflow"></a>

Conda is popular in data science because it manages Python packages plus system-level binaries.

Typical workflow:

```bash
conda create -n ds python=3.11 -y
conda activate ds
conda install numpy pandas matplotlib seaborn scikit-learn jupyterlab -y
python -c "import numpy, pandas; print(numpy.__version__, pandas.__version__)"
```


## 8. Common data science packages <a id="8-common-data-science-packages"></a>
