## 1) Why statements matter

Statements control program flow: what runs, how often it runs, and when it stops.
In data science and automation, statements are used for filtering rows, validating inputs, and looping through datasets.


```python
records = [
    {"name": "Ali", "score": 87},
    {"name": "Sara", "score": 42},
    {"name": "Mina", "score": 91},
]

# Example: statements decide which records are selected
for row in records:
    if row["score"] >= 50:
        print(row["name"], "passed")
```

## 2) `if` statement

Use `if` when a block should run only if a condition is true.
The condition must evaluate to a Boolean value (`True`/`False`).


```python
temperature = 29

if temperature > 30:
    print("It is a hot day")
```

```python
email = "student@gmail.com"

if "edu" in email:
    print("Email looks valid")
```

## 3) `if...else...elif`

- `else` handles the fallback case.
- `elif` lets you check multiple ordered conditions.

This pattern is common for grading, risk levels, and rule-based decision logic.


```python
score = 60

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "D"

print("grade:", grade)
```

```python
balance = 120
purchase = 150

if balance >= purchase:
    print("Payment approved")
else:
    print("Insufficient balance")
```

## 4) Boolean operators in conditions

Use `and`, `or`, and `not` to build richer rules.
- `and`: both conditions must be true
- `or`: at least one must be true
- `not`: reverses a Boolean result


```python
age = 22
has_ticket = True
is_banned = False

can_enter = age >= 18 and has_ticket and not is_banned
print("can_enter:", can_enter)
```

```python
city = "Boston"
weekend = False

if city == "Boston" and not weekend:
    print("Special local campaign is active")
```

## 5) Ternary expression

A ternary expression is a compact one-line `if...else`.
Use it for simple assignments, not for complex logic.


```python
num = 43
label = "high" if num > 100 else "low"
print(num, "is", label)
```

## 6) `for` loops

`for` loops iterate over items in a sequence (list, tuple, string, dictionary).
Use them when you know what to iterate over.


```python
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print(fruit)
```

```python
# Data scenario: compute average score
scores = [88, 92, 77, 81]
total = 0

for s in scores:
    total += s

print("average:", total / len(scores))
```

```python
# enumerate gives index + value
name = "python"
for i, ch in enumerate(name):
    print(i, ch)
```

### `range()` approaches in `for` loops

`range()` generates numbers and is commonly used when you need loop control.

Common forms:
- `range(stop)` -> starts at 0, ends before `stop`
- `range(start, stop)` -> starts at `start`, ends before `stop`
- `range(start, stop, step)` -> moves by `step` (can be negative)


```python
print("range(stop):", list(range(5)))
print("range(start, stop):", list(range(2, 7)))
print("range(start, stop, step):", list(range(1, 10, 2)))
print("reverse with negative step:", list(range(10, 0, -2)))
```

```python
items = ["a", "b", "c"]

# Approach 1: index-based loop with range + len
for i in range(len(items)):
    print(i, items[i])

# Approach 2 (often cleaner): enumerate
for i, value in enumerate(items):
    print(i, value)
```

## 7) `while` loops

`while` repeats as long as the condition stays true.
Use it when the number of iterations is unknown in advance.


```python
count = 1
while count <= 5:
    print("count:", count)
    count += 1
```

```python
# Retry scenario: stop when status becomes success
attempt = 1
status = "fail"

while attempt <= 3 and status != "success":
    print("attempt", attempt)
    if attempt == 3:
        status = "success"
    attempt += 1

print("final status:", status)
```

## 8) `break`, `continue`, `pass`

- `break`: stop the loop immediately
- `continue`: skip current iteration and move to next
- `pass`: placeholder (does nothing)


```python
# break example
for n in range(1, 10):
    if n == 5:
        break
    print(n)
```

```python
# continue example: print only odd numbers
for n in range(1, 8):
    if n % 2 == 0:
        continue
    print(n)
```

```python
# pass example
for task in ["load", "clean", "train"]:
    if task == "clean":
        pass  # TODO: implement cleaning step later     
    print("step:", task)
```

## 9) Sample


### Filter valid rows
Keep rows where age exists and is at least 18.


```python
rows = [
    {"name": "A", "age": 20},
    {"name": "B", "age": None},
    {"name": "C", "age": 16},
]

valid = []
for row in rows:
    if row["age"] is not None and row["age"] >= 18:
        valid.append(row)

print(valid)
```
