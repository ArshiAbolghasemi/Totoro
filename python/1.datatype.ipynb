## 1) Why data types matter

Data types affect memory usage, valid operations, and how calculations behave.
Using the right type makes your code safer and easier to debug.


```python
def inspect_value(x):
    print(f"value={x!r} | type={type(x).__name__}")

inspect_value(42)
inspect_value(3.14)
inspect_value("python")
```

## 2) Numeric types (`int`, `float`, `Decimal`)

- `int`: whole numbers
- `float`: decimal numbers (binary floating point)
- `Decimal`: exact base-10 decimals (useful for money)


```python
a = 15          # int
b = 2.5         # float

print(type(a), a)
print(type(b), b)
print("0.1 + 0.2 with float:", 0.1 + 0.2)
```

```python
from decimal import Decimal

print("Decimal add:", Decimal("0.1") + Decimal("0.2"))
price = Decimal("19.99")
tax = Decimal("0.07")
print("Total:", price * (Decimal("1") + tax))
```

## 3) Boolean (`bool`)

`bool` has two values: `True` and `False`. It is used in conditions and filters.


```python
age = 21
is_adult = age >= 18
print(is_adult)
```

## 4) String (`str`)

Strings represent text. They support indexing, slicing, and formatting.


```python
name = "Alice"
print(name.upper())
print(name[:3])
print(f"Hello, {name}!")
```

## 5) Sequence types (`list`, `tuple`)

- `list`: ordered and mutable
- `tuple`: ordered and immutable


```python
numbers = [10, 20, 30]
numbers.append(40)
print(numbers)

point = (3.5, 7.2)
print(point)
```

## 6) Set (`set`)

Sets store unique values and support fast membership tests.


```python
items = ["apple", "banana", "apple", "orange"]
unique_items = set(items)
print(unique_items)
print("banana" in unique_items)
```

## 7) Dictionary (`dict`)

Dictionaries map keys to values.


```python
student = {"name": "Sam", "age": 23, "city": "Boston"}
print(student["name"])
print(student.get("score", "N/A"))
```

## 8) `None` and missing values

`None` means "no value" and is common in data cleaning.


```python
income = None
if income is None:
    print("Income is missing")
```

## 9) Type conversion

Convert values explicitly when needed.


```python
raw_age = "29"
age = int(raw_age)
height = float("172.5")
print(age + 1)
print(height / 100)
```

## 10) Mutability basics

Mutable objects can change in place (`list`, `dict`, `set`).
Immutable objects cannot (`int`, `float`, `str`, `tuple`).


```python
text = "data"
# text[0] = "D"  # would fail because strings are immutable

values = [1, 2, 3]
values[0] = 100
print(values)
```
